```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data =pd.read_csv("https://raw.githubusercontent.com/SaiLinHtet-KiRA/MAT-252-Probability-Statistics/refs/heads/main/data/Data.csv")
data.columns =["Id","Gender","Age","Weight","Height","Blood","Favorite Number","Hometown","Birthday","Birth Date","Zodiac Sign"]

data[0:10]
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 5, figsize=(25, 10))

axs = axs.flatten()

for i, column in enumerate(data.drop(columns="Id").columns):
    data[column].value_counts().plot.bar(ax=axs[i])
    axs[i].set_ylabel('Count')


plt.tight_layout()

plt.show()
```
